```julia
#Definimos las funciones
f(z) = z^3-1
z0(z)=z-z/(1.0e10)
dercom(z)=(f(z)-f(z0(z)))/(z-z0(z)) 
```

```julia
#Definimos el método de Newton
function newtoncomp(z,l) 
    k=(z-(z+1)/1)
    if l ==1
        k
    else
    
        newtoncomp1(k,l-1)
    
    end
end

function newtoncomp1(z,l) 

        k=(z-f(z)/dercom(z))
    if l ==1
        k
    else
    
            newtoncomp1(k,l-1)
    
    end
end
```

```julia
ángulos=[1:1]
vi=1
vx=[vi*cos(θ) for θ in ángulos]
vy=[vi*sin(θ) for θ in ángulos]
```

```julia
function pm(ti, tf, xi, yi, vx, vy, theta)
    x(t,xi,vx)=xi+vx*t   
    y(t,yi,vy)=yi+vy*t-0.5*g*t^2
    g=9.81
    tiempos=[ti:0.5:tf]
    angulos=[0:pi/4:2pi]
    puntos_en_x=[x(t,xi,vx) for t in tiempos]
    puntos_en_y=[y(t,yi,vy) for t in tiempos]  
       
    
    velocidades_en_x=[vi*cos(theta) for theta in angulos]
    velocidades_en_y=[vi*sin(theta) for theta in angulos]
    
    tiempos, puntos_en_x, puntos_en_y, velocidades_en_x, velocidades_en_y
    end
```

```julia
tiempos,puntos_en_x,puntos_en_y,velocidades_en_x,velocidades_en_y=pm(0,9,0,99,1,1,pi)
```

```julia
plot(puntos_en_x,puntos_en_y)
xlabel(L"$x(t)$")
ylabel(L"$y(t)$")
```
